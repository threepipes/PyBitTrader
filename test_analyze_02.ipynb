```python
import sqlite3
import pandas as pd

db_name = 'db.sqlite3'
con = sqlite3.connect(db_name)
```

```python
df = pd.read_sql_query('select * from history limit 2000000, 4000000', con)
df
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
((df.price - df.price.shift()) / df.price).fillna(0).plot.hist(bins=100)
```

```python
diffs = ((df.price - df.price.shift()) / df.price).fillna(0)
diffs.std()
```

```python
diffs[(diffs != 0) & (diffs.abs() < 0.0025)].plot.hist(bins=100)
```

```python
diff_r = (df.price - df.price.shift())
diff_r.max(), diff_r.min()
```

```python
diff_r[diffs.abs() > 0.01].size
```

```python
df.exec_date = pd.to_datetime(df.exec_date)
df = df.set_index('exec_date')
df = df.loc['2000':]
df.price.plot()
```

```python
price_l = df.price.resample('5Min').mean()
price_l
```

```python
diff_l = (price_l - price_l.shift()) / price_l
diff_l.plot.hist(bins=1000, figsize=(10, 5))
```

```python
week_max = df.price.rolling(400000).max()
week_min = df.price.rolling(400000).min()
week_max
```

```python
dfh = df[['id', 'side', 'price', 'size']]
dfh['week_max'] = week_max
dfh['week_min'] = week_min
dfh
```

```python
dfh['week_max_per'] = dfh.week_max.shift(50000) / dfh.price - 1
dfh['week_min_per'] = dfh.week_min.shift(50000) / dfh.price - 1
```

```python
dfh.week_max_per.plot.hist(bins=100)
```

```python
dfh.week_min_per.plot.hist(bins=100)
```

```python
_max = dfh.week_max_per.sort_values().reset_index()
_max = _max[_max.notnull()]
_max = pd.Series(_max.index.values, index=_max.week_max_per)
_max
```

```python
(_max / _max.size).plot()
```

```python
_max /= _max.size
_max[_max < 0.3]
```

```python
_min = dfh.week_min_per.sort_values(ascending=False).reset_index()
_min = _min[_min.notnull()]
_min = pd.Series(_min.index.values, index=_min.week_min_per)
(_min / _min.size).plot()
```

```python
_min /= _min.size
_min[_min < 0.3]
```
