```python
orders = pd.read_sql('select * from orderdata', engine)
```

```python
orders
```

```python
real_order = orders[orders.child_order_id != '-']
double_order = real_order[real_order.side == real_order.side.shift(-1)]
double_order
```

```python
size_list = []
for timestamp in double_order.timestamp:
    timestamp -= datetime.timedelta(hours=9)
    from_time = timestamp - datetime.timedelta(seconds=20)
    to_time = timestamp + datetime.timedelta(seconds=20)
    pre_sell_size = sell_size[from_time:timestamp].sum()
    pre_buy_size = buy_size[from_time:timestamp].sum()
    post_sell_size = sell_size[timestamp:to_time].sum()
    post_buy_size = buy_size[timestamp:to_time].sum()
    side = 'SELL' if pre_sell_size > pre_buy_size else 'BUY '
    size_list.append({
        'timestamp': timestamp + datetime.timedelta(hours=9),
        'pre_sell': pre_sell_size,
        'pre_buy': pre_buy_size,
        'post_sell': post_sell_size,
        'post_buy': post_buy_size,
    })
    print(side, pre_sell_size, pre_buy_size)
```

```python
size_df = pd.DataFrame(size_list)
double_order[['side', 'price', 'timestamp']].merge(size_df, on='timestamp')
```
