```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
size = 1000
dn = np.random.randn(size)
p0 = 1000000
gwalk = np.cumprod(np.exp(dn*0.001))*p0
pd.Series(gwalk).plot(figsize=(10, 5))
```

```python
pd.Series(gwalk[600:800]).plot()
```

```python
def get_rw(size):
    dn = np.random.choice([-1,1], size=1000)
    swalk = np.cumsum(dn)
    return pd.Series(swalk)

def get_rwn(size):
    dn = np.random.randn(size)
    return pd.Series(np.cumsum(dn))
```

```python
get_rwn(10000).plot()
```

```python
rw = get_rwn(10000)
(rw - rw.shift()).plot(figsize=(10, 5))
```
